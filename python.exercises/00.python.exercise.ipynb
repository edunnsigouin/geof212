## Set up steps (you only need to do this once):

1. If you are off campus, login to the UiB VPN (Cisco).  

2. Open the terminal. 

    **Windows**: On windows 10, you can open the terminal by writing 'terminal' in the windows search bar. If your windows version is norwegian, you may also search for 'Ledertekst'. Finally, if you have anaconda installed on your personal computer, you may also use the windows search bar to find the 'anaconda prompt'. These are all different ways to access the terminal.

    **Linux or Mac**: open the terminal app.

3. In the terminal, log in to the remote machine called 'cyclone' by entering the command written below. Make sure to change the username to your UiB user name and press enter. Then enter your UiB password and press enter.

```
ssh your.user.name@cyclone.hpc.uib.no
```

4. In the terminal, download the latest version of the software manager 'miniconda' by entering:

```
wget https://repo.anaconda.com/miniconda/Miniconda3-py39_4.10.3-Linux-x86_64.sh ~
```

5. In the terminal, install miniconda by entering the command below. Follow the instructions by pressing 'enter' or writing 'yes'. 

    **This may take a few minutes**. 

```
sh ~/Miniconda3-py39_4.10.3-Linux-x86_64.sh
```

6. In the terminal, download the python materials for the course by entering the command below.

```
git clone https://github.com/edunnsigouin/geof212/ ~/geof212
```

7. In the terminal, create your python software environment using the .yml file you just downloaded by entering the command below.

    **This may take a few minutes**. 

```
conda env create --name geof212env --file=environment.yml
```

8. In the terminal, activate your python environment by typing the command written below:

```
conda activate geof212env
```

9. In the terminal, tell Jupyter lab that you want to use the python environment you just created by entering the command below:

```
python -m ipykernel install --user --name geof212env --display-name "geof212env"
```

10. In the terminal, start Jupyter lab by entering the command below. Some urls should pop up in the terminal. 

    **It's important that you choose a 4-digit number that is different from the one provided (1111). Only one person at a time can use a given port number and several of you will be using jupyter lab at once.** 

```
jupyter lab --no-browser --port=1111
```

11. Open up a new terminal and enter the command below. Then, enter your UiB password and press enter.

    **You must change the 4-digit number to the number you chose in the previous step**.

```
ssh -N -L 1111:localhost:1111 your.user.name@cyclone.hpc.uib.no
```

12. Open up a browser and enter one of the URLs given from step 10. Voila! Now you're ready to code in python on a remote supercomputer using the jupyter notebook on your internet browser.  

    To play around with the jupyter notebook, take a look at the notebook for this exercise by opening the file: 
    
    ```
    ~/geof212/exercises/00.python.exercise.ipynb
    ```
